## The Pythagorean Expectation for baseball

```python
# Here are the packages we need

import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Log of games played in 2018 from Retrosheet (https://www.retrosheet.org/)

MLB = pd.read_excel('MLB18.xlsx')
print(MLB.columns.tolist())
```

```python
# We can see what our dataframe looks like
MLB.head
```

```python
# For the Pythagorean Expectation we need only runs scored and conceded.

MLB18 = MLB[['VisitingTeam','HomeTeam','VisitorRunsScored','HomeRunsScore','Date']]
MLB18 = MLB18.rename(columns={'VisitorRunsScored':'VisR','HomeRunsScore':'HomR'})
MLB18
```

### Create columns to sum home and away wins

```python
# The variable 'hwin' is defined here as equaling 1 if the home team scored more runs, and zero otherwise. 
# The variable 'awin' is defined in a similar way for the away team.
# we also create a 'counter' variable = 1 for each row.

MLB18['hwin']= np.where(MLB18['HomR']>MLB18['VisR'],1,0)
MLB18['awin']= np.where(MLB18['HomR']<MLB18['VisR'],1,0)
MLB18['count']=1
MLB18
```

```python
# Here we define a df for home teams.
# to obtain the sum of wins and runs, and how many games were played 
 
MLBhome = MLB18.groupby('HomeTeam')[['hwin','HomR','VisR','count']].sum().reset_index()

MLBhome = MLBhome.rename(columns={'HomeTeam':'team','VisR':'VisRh','HomR':'HomRh','count':'Gh'})
MLBhome
```

```python
# Now we create a similar df for teams playing as visitors 
MLBaway = MLB18.groupby('VisitingTeam')[['awin','HomR','VisR','count']].sum().reset_index()
MLBaway = MLBaway.rename(columns={'VisitingTeam':'team','VisR':'VisRa','HomR':'HomRa','count':'Ga'})
MLBaway
```

```python
# We now merge MLBhome and MLBaway 
MLB18 = pd.merge(MLBhome,MLBaway,on='team')
MLB18
```

```python
# Now we create the total wins, games, played, runs scored and run conceded by summing the totals as home team and away team

MLB18['W']=MLB18['hwin']+MLB18['awin']
MLB18['G']=MLB18['Gh']+MLB18['Ga']
MLB18['R']=MLB18['HomRh']+MLB18['VisRa']
MLB18['RA']=MLB18['VisRh']+MLB18['HomRa']
MLB18
```

```python
# The last step is to define win percentage and the Pythagorean Expectation.

MLB18['wpc'] = MLB18['W']/MLB18['G']
MLB18['pyth'] = MLB18['R']**2/(MLB18['R']**2 + MLB18['RA']**2)
MLB18
```

```python
# Now it is nicely illustrated the close correlation between win percentage and the Pythagorean Expectation.
# The higher the Pythagorean Expectation, the more games a team is expected to win.

sns.lmplot(x='pyth', y='wpc', data=MLB18, fit_reg=False, hue='team',palette='viridis',legend=False)

plt.grid(False) 

plt.xlabel('Pythagorean Winning Percentage (pyth)')
plt.ylabel('Winning Percentage (wpc)')
plt.title('Pythagorean vs Winning Percentage')
plt.show()
```

```python
# Let´s do a Person and Spearman correlation between the Pythagorean Expectation and the Winning Percentage

pearson_corr = MLB18[['pyth','W']].corr(method='pearson')

spearman_corr = MLB18[['pyth','W']].corr(method='spearman')

print(pearson_corr)
print('           ')
print(spearman_corr)
```

```python
# Finally we generate a regression.

pyth_lm = smf.ols(formula = 'wpc ~ pyth', data=MLB18).fit()
pyth_lm.summary()
```

# OLS Regression Results Analysis

The regression output provides valuable insights into the relationship between win percentage (wpc) and the Pythagorean Expectation (pyth) in Major League Baseball.

## Coefficients Interpretation:

- The coefficient (coef) for the Pythagorean Expectation is 0.8770. This indicates that for every one-unit increase in the Pythagorean Expectation, the win percentage increases by 0.877.

## Statistical Significance:

- The standard error (std err) of the coefficient gives an indication of the precision of the estimate. The t statistic (t) assesses the statistical significance of the coefficient. A smaller p-value (P > |t|) indicates higher confidence in the relationship. In this case, the p-value is very small (below 0.05), suggesting a high level of confidence in the estimated relationship.

## Explained Variance:

- The R-squared value (0.894) indicates that approximately 89.4% of the variation in win percentage can be accounted for by the variation in the Pythagorean Expectation. This suggests that the Pythagorean Expectation is a strong predictor of win percentage in Major League Baseball.

## Overall Model Significance:

- The F-statistic (236.2) evaluates the overall significance of the model. The low p-value (3.54e-15) associated with the F-statistic indicates that the model as a whole is highly significant. This suggests that the Pythagorean Expectation is a significant predictor of win percentage in Major League Baseball.

## Adjusted R-squared:

- The Adjusted R-squared value (0.890) takes into account the number of predictors in the model. It indicates that approximately 89% of the variance in win percentage can be explained by the Pythagorean Expectation after adjusting for the number of predictors. This suggests that the model provides a good fit to the data.

In conclusion, the OLS regression results provide compelling evidence that the Pythagorean Expectation is a significant predictor of win percentage in Major League Baseball. The model demonstrates strong statistical significance and a high degree of explanatory power, suggesting that the Pythagorean Expectation is a reliable indicator of team success in baseball.


```python
# Plotting the regression line and the data points

sns.scatterplot(x="pyth", y="wpc", data=MLB18, hue='team', palette='viridis', legend=False)

# Plotting the regression line
plt.plot(MLB18['pyth'], pyth_lm.predict(MLB18['pyth']), color='navy', linewidth=2)
plt.grid(False) 

plt.xlabel('Pythagorean Winning Percentage (pyth)')
plt.ylabel('Winning Percentage (wpc)')
plt.title('Regression Analysis: Pythagorean vs Winning Percentage')

plt.show()

```
